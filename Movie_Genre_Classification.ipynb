#Removing Punctuation

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
  punctuations= string.punctuation
  return text.translate(str.maketrans('','',punctuations))
```

```python
df['clean_text']=df['clean_text'].apply(lambda x: remove_punctuation(x))
df.head()
```

#Removing **Stopwords**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
" , ".join(stopwords.words('english'))
```

```python
STOPWORDS= set(stopwords.words('english'))
def remove_stopwords(text):
  return" ".join([word for word in text.split() if word not in STOPWORDS])
```

```python
df['clean_text']=df['clean_text'].apply(lambda x: remove_stopwords(x))
df.head()
```

#Removal of Frequent Words

```python
from collections import Counter
word_count = Counter()
for text in df['clean_text']:
  for word in text.split():
    word_count[word] += 1

word_count.most_common(20)
```

```python
FREQUENT_Words = set(word for (word,wc) in word_count.most_common(10))
def remove_freq_words(text):
  return " ".join([word for word in text.split() if word not in FREQUENT_Words])
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_freq_words(x))
df.head()
```

#Removal of Rare words

```python
RARE_WORDS=set(word for(word,wc)in word_count.most_common()[:-10:-1])
RARE_WORDS
```

```python
def remove_rare_words(text):
  return" ".join([word for word in text.split() if word not in RARE_WORDS])
```

```python
import re
def cleaning1(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:cleaning1(x))
```

#Lemmatization and POS Tagging

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
Wordnet_map = {
    "N": wordnet.NOUN,
    "V": wordnet.VERB,
    "J": wordnet.ADJ,
    "R": wordnet.ADV
}

def lemmatize_words(text):
    pos_text = pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, Wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])
```

```python
nltk.download('averaged_perceptron_tagger_eng')
df['lemmatized_text'] = df['stemmed_text'].apply(lemmatize_words)
df.head()
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def correct_spellings(text):
    corrected_text = []
    misspelled_text = spell.unknown(text.split())

    for word in text.split():
        if word in misspelled_text:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)

    return " ".join(corrected_text)

```
